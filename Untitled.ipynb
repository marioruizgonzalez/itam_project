```python
from statsbombpy import sb
import pandas as pd
import matplotlib.pyplot as plt
from mplsoccer import Pitch

# Obtener los partidos de la Euro 2024
partidos_euro_2024 = sb.matches(competition_id=55, season_id=282)

# Selecciona el primer partido como ejemplo
partido_id = partidos_euro_2024.iloc[0]['match_id']

# Obtener los eventos del partido
eventos = sb.events(match_id=partido_id)

# Filtrar eventos de "Duel" (Duelo) donde el defensor es superado
duelos = eventos[eventos['type'] == 'Duel']
duelos_perdidos = duelos[duelos['duel_outcome'] == 'Lost']  # Filtrar los duelos perdidos

# Identificar en qué lado fueron desbordados (izquierdo o derecho)
# Si la posición X es menor a 60 se considera perfil izquierdo, mayor a 60, perfil derecho
duelos_perdidos['perfil_defensor'] = duelos_perdidos['location'].apply(lambda x: 'Izquierdo' if x[0] < 60 else 'Derecho')

# Contar cuántos desbordes ocurrieron por cada perfil
desbordes_por_perfil = duelos_perdidos['perfil_defensor'].value_counts()
print("Desbordes por perfil defensivo:\n", desbordes_por_perfil)
```

```python
import numpy as np
from statsbombpy import sb

# Obtener los partidos de la Euro 2024
partidos_euro_2024 = sb.matches(competition_id=55, season_id=282)

# Seleccionar un partido como ejemplo
partido_id = partidos_euro_2024.iloc[0]['match_id']

# Obtener los eventos del partido
eventos = sb.events(match_id=partido_id)

# Filtrar eventos de "Duel" (Duelo) donde el defensor es superado
duelos = eventos[eventos['type'] == 'Duel']
duelos_perdidos = duelos[duelos['duel_outcome'] == 'Lost']  # Filtrar los duelos perdidos

# Identificar en qué lado fueron desbordados (izquierdo o derecho)
duelos_perdidos['perfil_defensor'] = duelos_perdidos['location'].apply(lambda x: 'Izquierdo' if x[0] < 60 else 'Derecho')

# Contar cuántos desbordes ocurrieron por cada perfil
desbordes_por_perfil = duelos_perdidos['perfil_defensor'].value_counts()
print("Desbordes por perfil defensivo:\n", desbordes_por_perfil)

# Asegurarse de que los eventos tengan 'location' válido
duelos_perdidos = duelos_perdidos.dropna(subset=['location'])

# Calcular la distancia recorrida por el equipo rival tras superar al defensor
def calcular_ganancia_metros(row):
    siguiente_evento = eventos.loc[row.name + 1]  # Obtener el siguiente evento en el partido
    if pd.notna(siguiente_evento['location']) and pd.notna(row['location']):
        return np.sqrt((siguiente_evento['location'][0] - row['location'][0]) ** 2 +
                       (siguiente_evento['location'][1] - row['location'][1]) ** 2)
    else:
        return np.nan  # Si no hay 'location', devolver NaN

# Aplicar la función para calcular la ganancia de metros
duelos_perdidos['ganancia_metros'] = duelos_perdidos.apply(calcular_ganancia_metros, axis=1)

# Promedio de metros ganados por el equipo rival tras desbordar a los defensores
ganancia_promedio = duelos_perdidos['ganancia_metros'].mean()
print(f"Promedio de metros ganados tras superar a los defensores: {ganancia_promedio:.2f} metros")

```

```python
# Crear el campo de fútbol
pitch = Pitch(pitch_type='statsbomb', line_color='black')
fig, ax = pitch.draw(figsize=(10, 7))

# Extraer las posiciones de los duelos perdidos para visualizar las zonas vulnerables
x_coords = duelos_perdidos['location'].apply(lambda x: x[0])
y_coords = duelos_perdidos['location'].apply(lambda x: x[1])

# Visualización de las zonas donde los defensores son superados
pitch.scatter(x_coords, y_coords, ax=ax, color='red', label='Duelos Perdidos')

plt.title('Zonas Vulnerables - Duelos Perdidos')
plt.show()
```

```python
# Filtrar eventos peligrosos como tiros, pases clave y centros
peligro = eventos[(eventos['type'].isin(['Shot', 'Pass'])) & (eventos['pass_outcome'] == 'Completed')]

# Verificar si esos eventos ocurrieron en zonas vulnerables (dentro del área de los duelos perdidos)
peligros_en_zonas_vulnerables = peligro[
    (peligro['location'].apply(lambda x: any(np.sqrt((x[0] - x_duelo) ** 2 + (x[1] - y_duelo) ** 2) < 20
                                             for x_duelo, y_duelo in zip(x_coords, y_coords))))
]

# Contar cuántas jugadas peligrosas ocurrieron en zonas vulnerables
total_jugadas_peligrosas = len(peligros_en_zonas_vulnerables)
print(f"Total de jugadas peligrosas en zonas vulnerables: {total_jugadas_peligrosas}")
```
